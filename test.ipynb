```python
from solver import DuctSystem, DuctBranch, DuctMain, DuctNode, TerminalNode
```

```python
node_list = [DuctNode(0, (1,1)),
             DuctNode(1, (2,1)),
             DuctNode(2, (3,1)),
             DuctNode(3, (4,1)),
             DuctNode(4, (5,1))]
my_main = DuctMain(node_list)

my_terminal = TerminalNode(0, (5,5))

system = DuctSystem([my_terminal], my_main)
```

```python
a = system.all_geometry_coordinates()
```

```python
print(a)
```

```python
start_duct_node = system.main_duct_nodes[1]
objective_terminal = my_terminal

p = system.generate_path_to_terminal(start_duct_node, objective_terminal)
print(p)
```

```python
branch = system.generate_branch_from_path(p)
print([x.coordinate for x in branch.geometry])
```
